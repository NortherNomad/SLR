```python
import pandas as pd
import numpy as np
import csv
```

```python
file_path = "annotations.csv"
```

```python
trains = []
lengths = []
widths = []
heights = []
user_ids = []
texts = []
attachment_ids = []
args = []

with open(file_path, "r", newline="", encoding="utf-8") as file:
    reader = csv.reader(file, delimiter="\t")
    next(reader)
    for row in reader:
        attachment_id, text, user_id, height, width, length, train = row
        attachment_ids.append(attachment_id)
        texts.append(text)
        user_ids.append(user_id)
        heights.append(height)
        widths.append(width)
        lengths.append(length)
        trains.append(train)
```
